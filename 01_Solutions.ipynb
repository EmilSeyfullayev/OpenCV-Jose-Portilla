#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/content/drive/MyDrive/Computer-Vision-with-Python/DATA/dog_backpack.jpg'
image = cv2.imread(path, )
```

```python
plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
plt.imshow(cv2.flip(image_rgb, 1))
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
cv2.rectangle(image_rgb, pt1=(250, 400), pt2=(600, 700), color=(255, 0, 0), thickness=10)
plt.imshow(image_rgb)
```

```python
cv2.rectangle(image_rgb, pt1=(), pt2(), color=(255, 0, 0), thickness=10)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
pt1 = (400, 400)
pt2 = (250, 700)
pt3 = (600, 700)
```

```python
cv2.line(image_rgb, pt1, pt2, (0, 0, 255), 10)
cv2.line(image_rgb, pt2, pt3, (0, 0, 255), 10)
cv2.line(image_rgb, pt3, pt1, (0, 0, 255), 10)

plt.imshow(image_rgb)
```

```python
vertices = np.array([[250,700],[425,400],[600,700]],np.int32)
pts = vertices.reshape((-1,1,2))

```

```python
vertices
```

```python
cv2.polylines(image_rgb,[pts],isClosed=False,color=(255,0,0),thickness=20)
plt.imshow(image_rgb)
```

```python
cv2.fillPoly(image_rgb,[pts],color=(0,0,255))
plt.imshow(image_rgb)
```
